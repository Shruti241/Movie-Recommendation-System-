```python
import ast
#The 'ast' module is a built-in module in Python that provides tools for working with abstract syntax trees (ASTs). It is used to parse Python source code and represent it as a tree-like data structure.
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l+=[i['name']]
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert1(obj):
  l=[]
  count=0
  for i in ast.literal_eval(obj):
      l+=[i['name']]
      count+=1
      if(count==3):
          break;
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
def find_dir(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      l+=[i['name']]
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(find_dir)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new = movies[['movie_id','title','tag']]
```

```python
new['tag']=new['tag'].apply(lambda x:" ".join(x))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l1=[]
  for i in text.split():
    l1+=[ps.stem(i)]
  return " ".join(l1)
```

```python
new['tag']=new['tag'].apply(stem)
```

```python
new['tag']=new['tag'].apply(lambda x:x.lower())
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new['tag']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
```

```python
similarity_matrix
```

```python
sorted(list(enumerate(similarity_matrix[0])),reverse=True,key=lambda x:x[1])[1:6]
```

**Function to recommend the movie**

```python
from operator import index
def recommend(movie):
  index=new[new['title']==movie].index[0]
  distances=similarity_matrix[index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
